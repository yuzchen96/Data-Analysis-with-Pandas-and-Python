## Create a MultiIndex with `.set_index()` Method

```python
bigmac = pd.read_csv('bigmac.csv', parse_dates =['Date'])
bigmac.head()
```

```python
bigmac.set_index(['Date','Country'], inplace = True)
bigmac.head()
```

## Extract Index Level with `get_level_values` Method

```python
bigmac = pd.read_csv('bigmac.csv', parse_dates =['Date'], index_col = ['Date', 'Country'])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac.index.get_level_values('Date')
bigmac.index.get_level_values(0)
```

```python
bigmac.index.get_level_values('Country')
bigmac.index.get_level_values(1)
```

## Change Index Level Name with `set_names` Method

```python
bigmac = pd.read_csv('bigmac.csv', parse_dates =['Date'], index_col = ['Date', 'Country'])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac.index.set_names(['Day', 'Location'], inplace = True)
```

```python
bigmac.index
```

```python
bigmac.index.set_names(names = 'Date', level = 0)
bigmac.index.set_names(names = 'Date', level = 'Day', inplace = True)
bigmac.head()
```

## The `.sort_index()` Method on multiIndex DataFrame

```python
bigmac = pd.read_csv('bigmac.csv', parse_dates =['Date'], index_col = ['Date', 'Country'])
bigmac.head()
```

```python
bigmac.sort_index(ascending = [True,False], inplace = True)
bigmac.head()
```

```python
bigmac.sort_index(level = 0)
bigmac.sort_index(level = 'Date')
bigmac.sort_index(level = 1 )
bigmac.sort_index(level = 'Country', ascending = False)
```

## The `transpose` Method

```python
bigmac = pd.read_csv('bigmac.csv', parse_dates =['Date'], index_col = ['Date', 'Country'])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac = bigmac.transpose()
```

```python
bigmac.iloc[0]
```

```python
bigmac.loc[('Price in US Dollars',),('2010-01-01','China'):('2010-01-01','HongKong')]
```

##  The `swamplevel` Method

```python
bigmac = pd.read_csv('bigmac.csv', parse_dates =['Date'], index_col = ['Date', 'Country'])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac.swaplevel()
```

```python
bigmac.swaplevel('Date', 'Country')
bigmac.swaplevel('Country', 'Date')
bigmac.swaplevel(0, 1)
bigmac.swaplevel(1, 0)
```

```python
bigmac = bigmac.swaplevel(1, 0)
bigmac.head()
```

## The `.unstack()` Method

```python
world = pd.read_csv('worldstats.csv', index_col = ['country', 'year'])
world.head(3)
```

```python
s = world.stack()
s.head()
```

```python
s.unstack()
```

```python
s.head(3).unstack()
```

```python
s.unstack(2)
```

```python
s.unstack(0)
```

```python
s.unstack(1)
```

```python
s.unstack(-1)
```

```python
s.unstack('year')
```

```python
s.unstack('country')
```

```python
s.unstack(level = [1,0])
s.unstack(level = ['country','year'])
```

```python
s.unstack(level = [0,1])
s.unstack(level = ['year','country'])
```

```python
s.unstack('year', fill_value = 0)
```

## The `pivot` Method

```python
sales = pd.read_csv('salesmen.csv', parse_dates = ['Date'])
sales['Salesman'] = sales['Salesman'].astype('category')
sales.head()
```

```python
sales.pivot(index = 'Date', columns = 'Salesman', values = 'Revenue')
```

## The `pivot_table()` Method

```python
foods = pd.read_csv('foods.csv')
foods.head()
```

```python
foods.pivot_table(values = 'Spend', index = 'Gender', aggfunc = 'mean')
```

```python
foods.pivot_table(values = 'Spend', index = 'Gender', aggfunc = 'sum')
```

```python
foods.pivot_table(values = 'Spend', index = ['Gender', 'Item'], aggfunc = 'sum')
```

```python
foods.pivot_table(columns = 'City',values = 'Spend', index = ['Gender', 'Item'], aggfunc = 'sum')
```

```python
foods.pivot_table(columns = ['City', 'Frequency'],values = 'Spend', index = ['Gender', 'Item'], aggfunc = 'sum')
```

```python
foods.pivot_table(values = 'Spend', index= ['Gender', 'Item'], columns = 'City', aggfunc = 'count')
```
