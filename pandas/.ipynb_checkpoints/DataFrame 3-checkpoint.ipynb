## Review `.apply()` Method with columns

```python
movie = pd.read_csv('jamesbond.csv', index_col = 'Film')
movie.sort_index(inplace= True)
movie.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + ' Millions'
```

```python
columns = ['Box Office', 'Budget']

for column in columns:
    movie[column] = movie[column].apply(convert_to_string_and_add_millions)
```

```python
movie.head(3)
```

## The `.apply()` Method with row Values

```python
movie = pd.read_csv('jamesbond.csv', index_col = 'Film')
movie.sort_index(inplace= True)
movie.head(3)
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == 'Roger Morre':
        return 'Best'
    elif actor == 'Daniel Craig' and budget > 40:
        return 'Enjoyable'
    else:
        return 'I have no clue'
```

```python
movie['My Feelings'] = movie.apply(good_movie, axis = 1)
movie.head(3)
```
