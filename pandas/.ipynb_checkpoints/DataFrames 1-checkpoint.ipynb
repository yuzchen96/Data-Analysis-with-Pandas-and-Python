## Fill in null values with `.fillna()` Methods

```python
nba = pd.read_csv('nba.csv')
nba.head(3)
```

```python
nba['College'].fillna('No College', inplace = True)
nba.head(5)
```

## The `.astype()` Method

```python
nba = pd.read_csv('nba.csv').dropna(how = 'all')
nba.tail(5)
```

```python
# Requirement : No Null Values
nba['Salary'].fillna(0, inplace = True)
nba['College'].fillna('None', inplace = True)
nba.head(6)
```

```python
nba.dtypes
```

```python
nba.info()
```

```python
nba['Salary']=nba['Salary'].astype('int')
```

```python
nba.head(5)
```

```python
nba['Age']=nba['Age'].astype('int')
nba['Weight'] = nba['Weight'].astype('int')
```

```python
nba.head(5)
```

```python
nba['Position'].nunique()
```

```python
nba['Position'] = nba['Position'].astype('category')
```

```python
nba['Team'].nunique()
```

```python
nba['Team'] = nba['Team'].astype('category')
nba.info()
```

## Sort a DataFrame with the `.sort_values()` Method

```python
nba = pd.read_csv('nba.csv')
nba.head(3)
```

```python
nba.sort_values(by='Name', ascending = False)
```

```python
nba.sort_values(by = ['Name','Team'], ascending = [True, False])
```

## Sort DataFrame with `.sort_index()` Method

```python
nba = pd.read_csv('nba.csv')
nba.head(3)
```

```python
nba.sort_values(['Number','Salary','Name'], inplace = True)
nba.head(5)
```

```python
nba.sort_index()
```

## Rank Values with the `.rank()` Method

```python
nba = pd.read_csv('nba.csv').dropna(how = 'all')
nba['Salary']=nba['Salary'].fillna(0).astype('int')
nba.head(3)
```

```python
nba['Salary rank'] = nba['Salary'].rank(ascending = False).astype('int')
nba.head(3)
```

```python
nba.sort_values('Salary', ascending = False)
```
